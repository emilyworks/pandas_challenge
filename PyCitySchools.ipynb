```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "/Users/emilyye/BOOTCAMP WORK/TEMP/schools_complete.csv"
student_data_to_load = "/Users/emilyye/BOOTCAMP WORK/TEMP/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = len(school_data.index)

total_students = len(student_data.index)

total_budget = school_data["budget"].sum()

average_math_dist = student_data["math_score"].mean()

average_reading_dist = student_data["reading_score"].mean()

#math_passing_percent = students with a score above 70/total students
math_passing = len(student_data.loc[student_data["math_score"] >= 70, :])
math_passing_percent = math_passing/total_students

# reading_passing_percent
reading_passing = len(student_data.loc[student_data["reading_score"] >= 70, :])
reading_passing_percent = reading_passing/total_students

# math_reading_passing_percent
math_reading_passing = len(student_data.loc[(student_data["math_score"] >=70) & (student_data["reading_score"] >= 70), :])
math_reading_passing_percent = math_reading_passing/total_students

# districtSumm_df = pd.DataFrame()
district_summ_df = pd.DataFrame([
    {"Total Schools": total_schools,
    "Total Students": total_students,
    "Total Budget": total_budget,
    "Average Math Score": average_math_dist,
    "Average Reading Score": average_reading_dist,
    "Percent of Students Passing Math": math_passing_percent,
    "Percent of Students Passing Reading": reading_passing_percent,
    "Percent of Students Passing Math and Reading": math_reading_passing_percent}]
)

district_summ_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# make a copy of school_data_complete, which already holds some of the answers needed (school name, school type, total students, total school budget)
bySchool = school_data_complete

bySchool_gb = school_data_complete.groupby(["school_name"])

#calculate per student budget and add that to the bySchool DF
school_data["Per Student Budget"] = school_data["budget"]/school_data["size"]

bySchool_per_student = school_data.set_index(["school_name"])["Per Student Budget"]

# set the right indexes for non-calc'd answers (for answers that already exist in the original csv files)
bySchool_type = school_data.set_index(["school_name"])["type"]

bySchool_size = school_data.set_index(["school_name"])["size"]

bySchool_budget = school_data.set_index(["school_name"])["budget"]

# calculate percent passing math

filt_school_math = bySchool.loc[bySchool["math_score"]>=70,:]

passing_math = filt_school_math.groupby(["school_name"]).count()["math_score"]

students_per_school = bySchool.groupby(["school_name"]).count()["size"]

percent_passing_math = passing_math/students_per_school

# calculate percent passing reading
filt_school_reading = bySchool.loc[bySchool["reading_score"]>=70,:]

passing_reading = filt_school_reading.groupby(["school_name"]).count()["reading_score"]

percent_passing_reading = passing_reading/students_per_school

# calculate percent passing both

filt_school_both = bySchool.loc[(bySchool["reading_score"]>=70) & (bySchool["math_score"] >= 70), :]

passing_both = filt_school_both.groupby(["school_name"]).count()["reading_score"]

percent_passing_both = passing_both/students_per_school

#group by school to calculate average math, average reading

average_math = bySchool_gb["math_score"].mean()

average_reading = bySchool_gb["reading_score"].mean()

#combine everything into one dataframe

school_summ_df = pd.DataFrame(
     {"School Type": bySchool_type,
      "Total Students": bySchool_size,
      "Total Budget": bySchool_budget,
      "Budget Per Student": bySchool_per_student,
      "Average Math Score": average_math,
      "Average Reading Score": average_reading,
      "% Passing Math": percent_passing_math,
      "% Passing Reading": percent_passing_reading,
      "% Passing Both": percent_passing_both})

school_summ_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Define bins and labels

bins = [0,585,630,645,680]

labels = ["<585", "585-630","630-645","645-680"]

# making a copy of a previous DB and bininng by spending ranges per student
school_spending = school_summ_df
school_spending["Average Spending Range Per Student"] = pd.cut(school_spending["Budget Per Student"],bins, labels=labels)

# Groupby spending ranges

spending_scores = school_spending.groupby(["Average Spending Range Per Student"])

average_math_score_ss = spending_scores.mean()["Average Math Score"]
average_reading_score_ss = spending_scores.mean()["Average Reading Score"]
percent_passing_math_ss = spending_scores.mean()["% Passing Math"]
percent_passing_reading_ss = spending_scores.mean()["% Passing Reading"]
percent_passing_both_ss = spending_scores.mean()["% Passing Both"]

ss_summary = pd.DataFrame({
    "Average Math Score": average_math_score_ss,
    "Average Reading Score": average_reading_score_ss,
    "% Passing Math": percent_passing_math_ss,
    "% Passing Reading": percent_passing_reading_ss,
    "% Overall Passing": percent_passing_both_ss
}
)

ss_summary
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# define the bins and labels

bins_size = [0,1000, 2000, 5000]

labels_size = ["<1000","1000-2000","2000-5000"]

# define 
scores_size = school_summ_df
scores_size["School Size"] = pd.cut(scores_size["Total Students"],bins_size, labels=labels_size)

# Groupby size ranges and create a new DF based on that

scores_size_gb = scores_size.groupby(["School Size"])

average_math_score_size = scores_size_gb.mean()["Average Math Score"]
average_reading_score_size = scores_size_gb.mean()["Average Reading Score"]
percent_passing_math_size = scores_size_gb.mean()["% Passing Math"]
percent_passing_reading_size = scores_size_gb.mean()["% Passing Reading"]
percent_passing_both_size = scores_size_gb.mean()["% Passing Both"]

ss_summary_size = pd.DataFrame({
    "Average Math Score": average_math_score_size,
    "Average Reading Score": average_reading_score_size,
    "% Passing Math": percent_passing_math_size,
    "% Passing Reading": percent_passing_reading_size,
    "% Overall Passing": percent_passing_both_size
}
)

ss_summary_size
```
